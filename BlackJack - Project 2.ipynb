```python
# 52 cards
# bank account to place a bet
# computer one card face up and one card face down
# player two cards face up
# player has to be closer to 21 then dealer
# hit - new card (player)
# stay - no new card
# 2 turn dealer makesa hit until beat the player or bust
# bust - over 21
# face cards (jack, Queeen, King) = 10
# Ace is either 1 or 11
```

```python
import random

suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':10, 'Queen':10, 'King':10, 'Ace':11}
```

```python
class Cards:
    
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        
        self.value = values[rank]
        self.flip_side = 1
        
    def __str__(self):
        return f'{self.rank} of {self.suit}'
```

```python
class Player:
    
    def __init__(self,name, bet_money):
        self.name = name
        self.bet_money = bet_money
        self.player_cards = []
    
    def loose_money(self, amount):
        self.bet_money -= amount
    
    def win_money(self, amount):
        self.bet_money += amount
        
    def add_card(self, card):
        self.player_cards.append(card)
        
    def check_total(self):
        total = 0
        for card in self.player_cards:
            total += card.value
        
        return total
        
        
    
```

```python
class Computer:
    
    def __init__(self):
        self.name = 'Computer Player'
        self.computer_cards = []
    
    def add_card(self,card):
        self.computer_cards.append(card)
        
    def check_total(self):
        total = 0
        for card in self.computer_cards:
            total += card.value
        
        return total    
```

```python
class Deck:
    
    def __init__(self):
        self.all_cards = []
        for suit in suits:
            for rank in ranks:
                card = Cards(suit,rank)
                self.all_cards.append(card)
    
    def shuffle_cards(self):
        random.shuffle(self.all_cards)
    
    def remove_from_heap(self):
        card = self.all_cards.pop()
        return card
    
```

```python
# Set up game
run_game = True
game_on = False
while run_game:
    print('Would you like to play the game?(Y/N)')
    option = input()
    if option == 'Y':
        game_on = True
        name = input("Please provide your name: ")
        amount = ''
        while not isinstance(amount, float):
            try:
                amount = float(input("Please provide your amount money(must not be string): "))
            except:
                print('You must provide a number')
        player = Player(name,amount)    
        computer = Computer()
        deck = Deck()
        deck.shuffle_cards()

        for i in range(2):
            card = deck.remove_from_heap()
            player.add_card(card)
            card = deck.remove_from_heap()
            computer.add_card(card)  

        computer.computer_cards[1].flip_side = 0
        turn = 'player1'

    else:
        print('Ok, game ended')
        run_game = False
        break
        
    # Play the game    
    while game_on:
        # print cards
        print('Computer cards')
        for card in computer.computer_cards:
            if card.flip_side == 1:
                print(card, card.value)
            else:
                print('X')
        print('Player cards')
        for card in player.player_cards:
                print(card, card.value)
                
        # Player turn
        if turn == 'player1':
            print('Player turn')
            option = input('HIT or STAY: ')
            while option.upper() == 'HIT':
                card = deck.remove_from_heap()
                player.add_card(card)
                points = player.check_total()
                if points > 21:
                    print('Computer cards')
                    for card in computer.computer_cards:
                        print(card, card.value)
                    print('Player cards')
                    for card in player.player_cards:
                        print(card, card.value)
                    player.loose_money(amount)
                    print(player.bet_money)
                    print('Bust. You loose')
                    game_on = False 
                    break
                print('Computer cards')
                for card in computer.computer_cards:
                    if card.flip_side == 1:
                        print(card, card.value)
                else:
                    print('X')
                print('Player cards')
                for card in player.player_cards:
                    print(card, card.value)
                option = input('HIT or STAY')
                    
            if option.upper() == "STAY":
                print('Computer plays')
                turn = 'player2'
        
        # Computer turn
        if turn == 'player2':
            while computer.check_total() < 17:
                card = deck.remove_from_heap()
                computer.add_card(card)
                computer.computer_cards[-1].flip_side = 0
                comp_points = computer.check_total()
                if comp_points > 21:
                    print('Bust. Computer looses')
                    game_on = False 
            else:
                turn = 'check winner'
        
        # Check winner
        if turn == 'check winner':
            # print cards
            print('Computer cards')
            for card in computer.computer_cards:
                    print(card, card.value)
            print('Player cards')
            for card in player.player_cards:
                    print(card, card.value)
            if computer.check_total() > player.check_total():
                print('Computer wins')
                player.loose_money(amount)
                print(player.bet_money)
                game_on = False
                break
            elif computer.check_total() < player.check_total():
                print('You win')
                player.win_money(amount)
                print(player.bet_money)
                game_on = False
                break
            elif computer.check_total() == player.check_total():
                print('DRAW')
                game_on = False
                break
                
            


    
    
    
```
